## Reading the data from the csv file using pandas dataframe

```python
# importing pandas library
import pandas as pd
# reading csv file &  storing data in dataframe df
df = pd.read_csv("IPL 2022 Batters.csv")
df
```

## Handling null values

```python
# In 'Avg' column some cells have no numerical value.
# So we drop those rows (in a large dataset, dropping some rows shouldn't matter).
for i in range(162):
    if(df['Avg'][i] == '-'):
        df.drop(i,axis = 0, inplace = True)
df.reset_index(inplace = True, drop = True)
# Values stored in 'Avg' were srtings, so we add a new column to store these values as floats.
df['Avg_f'] = [float(x) for x in df['Avg']] 
df
```

```python
# Importing required functions from sklearn library
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import log_loss
```

# Linear Regression

```python
# checking how much each value relates to each other so that we can decide which columns to take
df.corr()
# note that Inns, Runs, BF and Avg_f have a good correlation with 4s
```

## Creating two dataframes one for features and one for target

```python
# X = dataframe for features and y = target dataframe
X = []
y = []
for i in range(154):
    X.append([df['Inns'][i], df['Runs'][i], df['BF'][i], df['Avg_f'][i]])
    y.append(df['4s'][i])
X, y
```

## Splitting the dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

## Training the model

```python
reg = LinearRegression().fit(X_train, y_train)
```

## Checking score, coefficients and intercept

```python
reg.score(X_train, y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

## Predicting values for train data and test data

```python
y_train_pred = reg.predict(X_train)
y_test_pred = reg.predict(X_test)
y_test_pred
```

## Calculating mean squared error

```python
mse(y_train, y_train_pred)
```

```python
mse(y_test, y_test_pred)
```

# Logistic Regression

```python
df
```

```python
# Creating another column named 'Performance' which stores number of 4s in binary form
# 0 if number of 4s is less than average number of 4s
# 1 if number of 4s is more than average number of 4s
temp = []
mean_4s = df['4s'].mean()
for i in range(154):
    if df['4s'][i] < mean_4s:
        temp.append(0)
    else :
        temp.append(1)
df['Performance'] = temp
```

```python
df
```

```python
# checking how much each value relates to each other so that we can decide which columns to take
df.corr()
# note that Inns, Runs, BF and Avg_f have a good correlation with Performance
```

## Creating two dataframes one for features and one for target

```python
X = []
y = []
for i in range(154):
    X.append([df['Inns'][i], df['Runs'][i], df['BF'][i], df['Avg_f'][i]])
    y.append(df['Performance'][i])
X, y
```

## Splitting the dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

## Training the model

```python
logr = LogisticRegression().fit(X_train, y_train)
```

## Checking score, coefficients and intercept

```python
logr.score(X_train, y_train)
```

```python
logr.coef_
```

```python
logr.intercept_
```

## Predicting values for train data and test data

```python
y_train_pred = logr.predict(X_train)
y_test_pred = logr.predict(X_test)
y_test_pred
```

## Calculating mean squared error

```python
mse(y_train, y_train_pred)
```

```python
mse(y_test, y_test_pred)
```

## Calculating log loss

For Logistic Regression most popular loss function is Log Loss.

```python
log_loss(y_train, y_train_pred)
```

```python
log_loss(y_test, y_test_pred)
```
